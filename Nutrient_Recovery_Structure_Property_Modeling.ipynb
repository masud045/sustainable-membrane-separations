```python
# Logistic growth model
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Experimental data
x_data = np.array([0, 0.5, 1.5, 2.5])  # GA (wt%) concentrations
y_data = np.array([61.92, 33.45, 30.6, 27.948])  # Ammonium permeance values

# Define the logistic function L / (1 + exp(-k * (x - x0)))
def logistic_func(x, L, k, x0):
    return L / (1 + np.exp(-k * (x - x0)))

# Fit the model to the data
params, _ = curve_fit(logistic_func, x_data, y_data, p0=[60, 1, 1], maxfev=10000)  # Initial guess for L, k, x0

# Extract fitted parameters
L, k, x0 = params
print(f"Fitted parameters: L = {L}, k = {k}, x0 = {x0}")

# Generate continuous x values for the plot
x_continuous = np.linspace(min(x_data), max(x_data), 100)
y_continuous = logistic_func(x_continuous, L, k, x0)

# Plot the experimental data
plt.scatter(x_data, y_data, color='red', label='Experimental data')

# Plot the fitted curve
plt.plot(x_continuous, y_continuous, label=f'Logistic Fit: L / (1 + exp(-k(x - x0)))', color='blue')

# Add labels and legend
plt.xlabel('GA concentration (wt%)')
plt.ylabel('Ammonium Permeance')
plt.title('Logistic Growth Fit of Ammonium Permeance vs GA Concentration')
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Experimental data for potassium
x_data_k = np.array([0, 0.5, 1.5, 2.5])  # GA (wt%) concentrations
y_data_k = np.array([102, 40, 30, 31])   # Potassium permeance values

# Experimental data for ammonium
x_data_nh4 = np.array([0, 0.5, 1.5, 2.5])  # GA (wt%) concentrations
y_data_nh4 = np.array([152, 92, 81, 58])   # Ammonium permeance values

# Define the rational function: (a * x + b) / (x + c)
def rational_func(x, a, b, c):
    return (a * x + b) / (x + c)

# Fit for potassium
params_potassium, _ = curve_fit(rational_func, x_data_k, y_data_k, p0=[1, 1, 1])
a_k, b_k, c_k = params_potassium
print(f"Potassium Fit: a = {a_k:.4f}, b = {b_k:.4f}, c = {c_k:.4f}")

# Fit for ammonium
params_ammonium, _ = curve_fit(rational_func, x_data_nh4, y_data_nh4, p0=[1, 1, 1])
a_nh4, b_nh4, c_nh4 = params_ammonium
print(f"Ammonium Fit: a = {a_nh4:.4f}, b = {b_nh4:.4f}, c = {c_nh4:.4f}")

# Generate continuous GA concentration values
x_continuous = np.linspace(0, 2.5, 200)

# Get predicted permeance values
y_k_pred = rational_func(x_continuous, *params_potassium)
y_nh4_pred = rational_func(x_continuous, *params_ammonium)

# Plot Potassium fit
plt.figure()
plt.scatter(x_data_k, y_data_k, color='red', label='Experimental Data')
plt.plot(x_continuous, y_k_pred, color='blue', label='Rational Fit')
plt.xlabel('GA concentration (wt%)')
plt.ylabel('Potassium Permeance')
plt.title('Potassium Permeance vs GA Concentration')
plt.legend()
plt.show()

# Plot Ammonium fit
plt.figure()
plt.scatter(x_data_nh4, y_data_nh4, color='red', label='Experimental Data')
plt.plot(x_continuous, y_nh4_pred, color='blue', label='Rational Fit')
plt.xlabel('GA concentration (wt%)')
plt.ylabel('Ammonium Permeance')
plt.title('Ammonium Permeance vs GA Concentration')
plt.legend()
plt.show()

# Calculate R² for Potassium
y_k_actual_pred = rational_func(x_data_k, *params_potassium)
sst_k = np.sum((y_data_k - np.mean(y_data_k)) ** 2)
ssr_k = np.sum((y_data_k - y_k_actual_pred) ** 2)
r2_k = 1 - ssr_k / sst_k
print(f"Potassium R²: {r2_k:.4f}")

# Calculate R² for Ammonium
y_nh4_actual_pred = rational_func(x_data_nh4, *params_ammonium)
sst_nh4 = np.sum((y_data_nh4 - np.mean(y_data_nh4)) ** 2)
ssr_nh4 = np.sum((y_data_nh4 - y_nh4_actual_pred) ** 2)
r2_nh4 = 1 - ssr_nh4 / sst_nh4
print(f"Ammonium R²: {r2_nh4:.4f}")

# Compute selectivity (Ammonium / Potassium)
selectivity = y_nh4_pred / y_k_pred

# Find the maximum selectivity and corresponding GA concentration
max_index = np.argmax(selectivity)
max_selectivity = selectivity[max_index]
optimal_ga = x_continuous[max_index]

# Plot Selectivity
plt.figure()
plt.plot(x_continuous, selectivity, color='green', label='Selectivity (NH₄⁺ / K⁺)')
plt.axvline(optimal_ga, color='gray', linestyle='--', label=f'Max at {optimal_ga:.2f} wt%')
plt.xlabel('GA concentration (wt%)')
plt.ylabel('Selectivity (Ammonium / Potassium)')
plt.title('Selectivity vs GA Concentration')
plt.legend()
plt.show()

print(f"Maximum selectivity: {max_selectivity:.4f} at GA concentration: {optimal_ga:.2f} wt%")
```
